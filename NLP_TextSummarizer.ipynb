# Method 3: SpaCy

**How to do text Summarization:**
* Text Cleaning
* Sentence Tokenization
* Word Tokenization
* Word Frequency Table
* Summarization

```python
!pip install -U spacy
!python -m spacy download en_core_web_sm
```

```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
```

```python
stopwords=list(STOP_WORDS)
stopwords
```

```python
nlp=spacy.load("en_core_web_sm")
```

```python
doc=nlp(text)
```

```python
doc
```

```python
tokens=[token.text for token in doc]
```

```python
tokens
```

```python
punctuation
```

```python
punctuation=punctuation+'\n'
```

```python
word_freq={}
for word in doc:
  if word.text.lower() not in stopwords:
    if word.text.lower() not in punctuation:
      if word.text not in word_freq.keys():
        word_freq[word.text]=1
      else:
        word_freq[word.text] +=1
```

```python
word_freq
```

```python
max_freq=max(word_freq.values())
max_freq
```

# Normalizing the frequencies

```python
for word in word_freq.keys():
  word_freq[word]=word_freq[word]/max_freq
```

```python
word_freq
```

```python
sentence_tokens=[sent for sent in doc.sents]
sentence_tokens
```

```python
sentences_scores={}
for sent in sentence_tokens:
  for word in sent:
    if word.text.lower() in word_freq.keys():
      if sent not in sentences_scores.keys():
        sentences_scores[sent]=word_freq[word.text.lower()]
      else:
        sentences_scores[sent]+=word_freq[word.text.lower()]
```

```python
sentences_scores
```

```python
from heapq import nlargest
```

```python
select_length=int(len(sentence_tokens)*0.3)
select_length
```

```python
summary=nlargest(select_length,sentences_scores,key=sentences_scores.get)
```

```python
summary
```

```python
final_sum=[word.text for word in summary]
```

```python
summary=' '.join(final_sum)
```
